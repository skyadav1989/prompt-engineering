# 🚀 Phase 2 — Advanced Prompting Techniques (Practical)

This notebook demonstrates practical implementations of:
- Chain of Thought (CoT)
- Few-shot prompting
- Self-consistency
- Structured output
- Prompt chaining

Using Gemini models.

```python
!pip install -q google-generativeai
import google.generativeai as genai
from google.colab import userdata



genai.configure(api_key=userdata.get('GEMINI_API_KEY'))
model = genai.GenerativeModel("gemini-2.5-flash-lite")
```

## 1️⃣ Chain of Thought (Step-by-Step Reasoning)

```python
prompt = """
Solve step by step:
If a product costs ₹2000 and discount is 18%, what is final price?
Show calculations clearly before final answer.
"""

response = model.generate_content(prompt, generation_config={"temperature": 0.3})
print(response.text)
```

## 2️⃣ Few-Shot Prompting (Pattern Learning)

```python
prompt = """
Classify sentiment:

Review: The product is terrible.
Sentiment: Negative

Review: Excellent build quality.
Sentiment: Positive

Review: Cooling performance is average.
Sentiment:
"""

response = model.generate_content(prompt, generation_config={"temperature": 0.2})
print(response.text)
```

## 3️⃣ Self-Consistency (Multiple Candidates)

```python
response = model.generate_content(
    "Solve step by step: 36 × 27",
    generation_config={"temperature": 0.8, "candidate_count": 3}
)

for i, candidate in enumerate(response.candidates):
    print(f"Candidate {i+1}:")
    print(candidate.content.parts[0].text)
    print("-" * 40)
```

## 4️⃣ Structured JSON Output (API Style)

```python
prompt = """
Return ONLY valid JSON.
{
  "title": string,
  "meta_description": string,
  "keywords": list
}

Product: 1.5 Ton Inverter AC
"""

response = model.generate_content(
    prompt,
    generation_config={"temperature": 0.3, "max_output_tokens": 150}
)

print(response.text)
```

## 5️⃣ Prompt Chaining (Multi-Step System Simulation)

```python
# Step 1: Intent Classification
intent_prompt = """
Classify user intent:
Options: Product Inquiry, Complaint, Technical Issue
Message: My AC is not cooling properly.
"""

intent_response = model.generate_content(intent_prompt, generation_config={"temperature": 0.2})
print("Intent:", intent_response.text)

# Step 2: Generate Response
intent_text = intent_response.text.strip()
response_prompt = f"""
User intent detected: {intent_text}

Generate troubleshooting steps in bullet format.
"""

final_response = model.generate_content(response_prompt, generation_config={"temperature": 0.4})
print("\nSupport Response:")
print(final_response.text)
```
